# Normalization

#### Manually

```python
minOf_X = df['Price'].min()
```

```python
minOf_X
```

```python
maxOf_X = df['Price'].max()
```

```python
maxOf_X
```

```python
df['price_norm_manual'] = (df['Price'] - minOf_X) / (maxOf_X - minOf_X)
```

```python
df
```

#### Using Sklearn <b>

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
```

```python
df['price_norm_sklearn'] = mms.fit_transform(df[['Price']])
```

```python
df
```

# Standardization

#### Manually

```python
mean_x = np.mean(df['Price'])
```

```python
mean_x
```

```python
std = df['Price'].std(ddof=0) #not N-1
```

```python
std
```

```python
df['price_stand_manual'] = (df['Price'] - mean_x) / std
```

```python
df
```

#### Using sklearn

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
df['price_stand_sklearn'] = std_scaler.fit_transform(df[['Price']])
```

```python
df
```

# Log Transformation

#### Manually with numpy

```python
df['price_logTrans_manual'] = np.log1p(df['Price']) 
```

```python
df
```

#### Using Sklearn

```python
from sklearn.preprocessing import FunctionTransformer
logTrans = FunctionTransformer(np.log1p)
```

```python
df['Price_logTrans_sklearn'] = logTrans.fit_transform(df[['Price']])
```

```python
df
```

# Max Absolute Scaler

#### Manually

```python
mxABS = df['Price'] / np.abs(np.max(df['Price']))
```

```python
df['price_MABS_manual'] = mxABS
```

```python
df
```

#### Using Sklearn

```python
from sklearn.preprocessing import MaxAbsScaler
mxABS2 = MaxAbsScaler()
df['price_MABS_sklearn'] = mxABS2.fit_transform(df[['Price']])
```
